```python
import requests
import json 
import pandas as pd
```

```python
model = 'grand cherokee'
api_url = 'https://api.api-ninjas.com/v1/cars?limit=50&year=2010&model={}'.format(model)
response = requests.get(api_url, headers={'X-Api-Key': 'ZCdETropEXWOtXAjrg7ceQ==jfg7IuXZLUUuwjMi'})

cherokee2010_json = json.loads(response.text)
cherokee2010_df = pd.json_normalize(cherokee2010_json)

cherokee2010_df
```
